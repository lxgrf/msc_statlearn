```python
import pandas as pd
```

```python
trainraw = pd.read_csv("train_selected.csv")
trainraw.head()
```

```python
# ALl up stats
trainraw.describe()
```

```python
#All up dataset histograms
for column in trainraw: trainraw.hist(column=column,  
                                      bins=50)
```

```python
# Plot all columns for dataset 1
for column in trainraw:
    trainraw[trainraw.id == 1].plot.line(y=column)
```

```python
# Normalising - for ease of plotting!
for a in ["s1", "s2", "s3", "s4"]:
    cmin = trainraw[a].min()
    cdiff = trainraw[a].max() - cmin
    trainraw["{}_norm".format(a)] = (trainraw[a] - cmin) / cdiff

trainraw.head()
```

```python
# Normalised plots
for i in range(1,11):
    trainraw[trainraw.id == i][["s1_norm","s2_norm","s3_norm","s4_norm","label_bnc"]].plot.line()
```
